### 3. (1번 산출물을 활용하여) 년도별 행복지수 평균이 유의미하게 차이가 나는지 알아보고자 한다. 이와 관련하여 적절한 검정을 사용하고 검정통계량을 기술하시오.<br>- 해당 검정의 검정통계량은 자유도가 2인 F 분포를 따른다.<br>- 검정통계량은 소수점 넷째 자리까지 기술한다.<br>(답안 예시) 0.1234

```python
# 3개 이상의 표본 → ANOVA
```

```python
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
```

```python
q3 = dataset11[dataset11['Country'].isin(q1_2.index.values)]

q3
```

```python
anova = ols('Happiness_Score~C(year)', q3).fit()   # C(수치형 변수) : 강제적으로 범주형으로 인식
# ols 사용 시 변수명에서 스페이스, -, .이 포함된 변수는 사용 불가
```

```python
anova_lm(anova)
```

```python
# C(year)  : 집단 간
# Residual : 집단 내
# PR(>F)   : p-value   →   0.9957 > 0.05이므로 귀무가설 기각 X
```

```python
# 적어도 그룹 i는 평균이 같지 않다.
```

#### 정답 : ***0.0042***

```python
##### ols를 쓰면 다음이 가능! #####
from statsmodels.stats.multicomp import pairwise_tukeyhsd   # 다중 비교(사후분석)
```

```python
q3_out = pairwise_tukeyhsd(q3['Happiness_Score'], q3['year'])

print(q3_out)
```

```python
pos.columns
```

```python
top10 = pos['itemDescription'].value_counts().nlargest(10).index

top10
```

```python
q3 = q2[q2['itemDescription'].isin(top10)]

q3
```

```python
q3_tab = pd.pivot_table(data=q3, index=['Date', 'day'],
                        values='itemDescription',
                        aggfunc='count').reset_index()

q3_tab
```

```python
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
```

```python
ols1 = ols('itemDescription~day', q3_tab).fit()
q3_out = anova_lm(ols1)

q3_out
```

```python
q3_out['PR(>F)'][0]
```

#### 정답 : ***0.52***
