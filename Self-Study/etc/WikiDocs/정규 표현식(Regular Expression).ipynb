## **2. 정규 표현식 실습**

### **1) . 기호**

```python
import re
r = re.compile("a.c")
r.search("kkk")   # 아무런 결과도 출력되지 않는다.
```

```python
r.search("abc")
```

### **2) ? 기호**

```python
import re
r = re.compile("ab?c")
r.search("abbc")   # 아무런 결과도 출력되지 않는다.
```

```python
r.search("abc")
```

```python
r.search("ac")
```

### **3) * 기호**

```python
import re
r = re.compile("ab*c")
r.search("a")   # 아무런 결과도 출력되지 않는다.
```

```python
r.search("ac")
```

```python
r.search("abc")
```

```python
r.search("abbbbc")
```

### **4) + 기호**

```python
import re
r = re.compile("ab+c")
r.search("ac")   # 아무런 결과도 출력되지 않는다.
```

```python
r.search("abc")
```

```python
r.search("abbbbc")
```

### **5) ^ 기호**

```python
import re
r=re.compile("^a")
r.search("bbc")   # 아무런 결과도 출력되지 않는다.
```

```python
r.search("ab")
```

### **6) {숫자} 기호**

```python
import re
r=re.compile("ab{2}c")
r.search("ac")    # 아무런 결과도 출력되지 않는다.
r.search("abc")   # 아무런 결과도 출력되지 않는다.
```

```python
r.search("abbc")
```

```python
r.search("abbbbbc")   # 아무런 결과도 출력되지 않는다.
```

### **7) {숫자1, 숫자2} 기호**

```python
import re
r=re.compile("ab{2,8}c")
r.search("ac")    # 아무런 결과도 출력되지 않는다.
r.search("abc")   # 아무런 결과도 출력되지 않는다.
```

```python
r.search("abbc")
```

```python
r.search("abbbbbbbbc")
```

```python
r.search("abbbbbbbbbc")   # 아무런 결과도 출력되지 않는다.
```

### **8) {숫자,} 기호**

```python
import re
r=re.compile("a{2,}bc")
r.search("bc")   # 아무런 결과도 출력되지 않는다.
r.search("aa")   # 아무런 결과도 출력되지 않는다.
```

```python
r.search("aabc")
```

```python
r.search("aaaaaaaabc")
```

### **9) [ ] 기호**

```python
import re
r=re.compile("[abc]")   # [abc]는 [a-c]와 같다.
r.search("zzz")         # 아무런 결과도 출력되지 않는다.
```

```python
r.search("a")
```

```python
r.search("aaaaaaa")
```

```python
r.search("baac")
```

```python
import re
r=re.compile("[a-z]")
r.search("AAA")   # 아무런 결과도 출력되지 않는다.
```

```python
r.search("aBC")
```

```python
r.search("111")   # 아무런 결과도 출력되지 않는다.
```

### **10) [^문자] 기호**

```python
import re
r=re.compile("[^abc]")
r.search("a")    # 아무런 결과도 출력되지 않는다.
r.search("ab")   # 아무런 결과도 출력되지 않는다.
r.search("b")   # 아무런 결과도 출력되지 않는다.
```

```python
r.search("d")
```

```python
r.search("1")
```

## **3. 정규 표현식 모듈 함수 예제**

### **(1) re.match()와 re.search()의 차이**

```python
import re
r = re.compile("ab.")
```

```python
r.search("kkkabc")
```

```python
r.match("kkkabc")   # 아무런 결과도 출력되지 않는다.
```

```python
r.match("abckkk")
```

### <span style="color:red">**(2) re.split()**</span>

```python
import re
text = "사과 딸기 수박 멜론 바나나"
re.split(" ", text)
```

```python
import re
text = """사과
딸기
수박
멜론
바나나"""
re.split("\n", text)
```

```python
import re
text = "사과+딸기+수박+멜론+바나나"
re.split("\+", text)
```

### **(3) re.findall()**

```python
import re
text = """이름 : 김철수
전화번호 : 010 - 1234 - 1234
나이 : 30
성별 : 남"""
re.findall("\d+", text)
```

```python
re.findall("\d+", "문자열입니다.")
```

### **(4) re.sub()**

```python
import re
text = "Regular expression : A regular expression, regex or regexp[1] (sometimes called a rational expression)[2][3] is, in theoretical computer science and formal language theory, a sequence of characters that define a search pattern."
re.sub("[^a-zA-Z]", " ", text)
```

## **5. 정규 표현식 텍스트 전처리 예제**

```python
import re

text = """100 John    PROF
101 James   STUD
102 Mac   STUD"""

re.split("\s+", text)
```

```python
re.findall("\d+", text)
```

```python
re.findall("[A-Z]{4}", text)
```

```python
re.findall("[A-Z][a-z]+", text)
```

## **6. 정규 표현식을 이용한 토큰화**

```python
import nltk
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer("[\w]+")
print(tokenizer.tokenize("Don't be fooled by the dark sounding name, Mr. Jone's Orphanage is as cheery as cheery goes for a pastry shop"))
```

```python
import nltk
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer("[\s]+", gaps=True)
print(tokenizer.tokenize("Don't be fooled by the dark sounding name, Mr. Jone's Orphanage is as cheery as cheery goes for a pastry shop"))
```
